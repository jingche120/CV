# 爬Learn Code With Mike 最近貼文

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

# 此處的瀏覽器設定就是不要有彈出視窗
options = Options()
options.add_argument("--disable-notifications")
#傳入剛剛所下載的「瀏覽器驅動程式路徑」及「瀏覽器設定(chrome_options)」
driver = webdriver.Chrome('./chromedriver',chrome_options=options)
# # 連入FB，並登入我的帳號
driver.get('https://www.facebook.com/')
# 因為改版，所要email = driver.find_element_by_id("email")要改成下面的
email = driver.find_element(By.ID, "email")
password = driver.find_element(By.ID, "pass")
login = driver.find_element(By.NAME, "login")
email.send_keys('sinonno1006@yahoo.com.tw')
password.send_keys('qsx123')
login.click()
time.sleep(5)
# 登入成功
#進入Learn Code With Mike粉絲專頁
driver.get('https://www.facebook.com/learncodewithmike')
time.sleep(5)
#執行滾動捲軸3次，讀入更多文章
for x in range(1, 4):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

unprocessedDatas = soup.find_all('div', {
    'class':'x1iorvi4 x1pi30zi x1swvt13 x1l90r2v'
})
content = []
for unprocessedData in unprocessedDatas: 
    post = unprocessedData.find('span', {'dir': 'auto'}) 
    if post:
        content.append(post.getText())

import os
file = open('爬粉專最新貼文標題.txt','w+', encoding='UTF-8')
for i in content:
    file.write(i)
    file.write('\n')
file.close()
driver.quit()    
```
